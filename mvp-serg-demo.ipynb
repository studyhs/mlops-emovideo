```python
from facial_emotion_recognition import EmotionRecognition
import cv2
import os
```

```python
er = EmotionRecognition(device='cpu')
```

```python
emo_images_folder = 'emo_images'
```

```python
if not os.path.exists(emo_images_folder):
    os.mkdir(emo_images_folder)
```

```python
video_capture = cv2.VideoCapture("./videos/serg-only.mp4")
```

```python
i = 1
while True:
    ret, frame = video_capture.read()

    if not ret:
        break
        
    text = ""
    detect = True 
    
    detect, frame = er.recognise_emotion(frame, return_type='BGR')
    
    if detect:
        cv2.putText(frame, text, (90, 60), cv2.FONT_HERSHEY_DUPLEX, 1.6, (147, 58, 31), 2)
        cv2.imwrite(f'{emo_images_folder}/frame-{i:03}.jpg', frame)
    
        i += 1
    
    if cv2.waitKey(1) == 27:
        break
        
video_capture.release()
cv2.destroyAllWindows()        
```
